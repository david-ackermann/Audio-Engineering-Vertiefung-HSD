## 1. Mathematische Grundlage

Die kontinuierliche Fourier-Transformation untersucht, wie stark ein Signal `x(t)` zu einer komplexen Schwingung der Frequenz `f` passt:

$$
X(f)=\int x(t)\,e^{-j2\pi f t}\,\mathrm{d}t
$$

Der komplexe Exponentialterm kann mit der Euler-Beziehung zerlegt werden:

$$
e^{-j2\pi f t}=\cos(2\pi f t)-j\sin(2\pi f t)
$$

Damit wird die Analysegleichung zu

$$
X(f)=\int x(t)\cos(2\pi f t)\,\mathrm{d}t
-j\int x(t)\sin(2\pi f t)\,\mathrm{d}t
$$

Didaktisch ist das sehr wichtig:

- der **Cosinus-Anteil** liefert die Projektion auf die reelle Achse,
- der **Sinus-Anteil** liefert die Projektion auf die imaginäre Achse,
- beide zusammen ergeben den komplexen Fourier-Koeffizienten `X(f)`.


## 2. Was bedeutet das anschaulich?

Die Fourier-Analyse stellt für jede geprüfte Frequenz eine rotierende Referenzschwingung bereit. Wenn Signal und Referenz gut zusammenpassen, addieren sich die Produkte über die Zeit konstruktiv. Wenn sie nicht zusammenpassen, heben sich positive und negative Beiträge weitgehend auf.

Der resultierende komplexe Wert kann als Zeiger in der komplexen Ebene gelesen werden:

- die **Länge** des Zeigers entspricht dem **Betrag** `|X(f)|`,
- der **Winkel** des Zeigers entspricht der **Phase** von `X(f)`.

So entsteht aus der Integralgleichung direkt eine geometrische Interpretation.

## 3. Beobachtungsdauer und Fensterung

In der Praxis wird ein Signal nur über ein endliches Zeitfenster beobachtet. In diesem Notebook wird deshalb mit einer Fensterfunktion `w(t)` gearbeitet:

$$
X_{T_\mathrm{obs},w}(f)=\int_{-T_\mathrm{obs}/2}^{T_\mathrm{obs}/2} x(t)\,w(t)\,e^{-j2\pi f t}\,\mathrm{d}t
$$

Daraus folgen zwei zentrale Beobachtungen:

- Eine **längere Beobachtungsdauer** führt zu **schmaleren Spektralpeaks**.
- Die **Fensterfunktion** beeinflusst Nebenkeulen und Peakform im Spektrum.

Mit dem Signalmodus `Constant` kann die Wirkung der Fensterfunktion besonders gut isoliert betrachtet werden, weil dann direkt das fenstergewichtete Signal analysiert wird.

## 4. Aufbau der interaktiven Abbildung

Die App zeigt dieselbe Fourier-Analyse gleichzeitig in mehreren Darstellungen:

1. **Zeitbereich links oben**: Signal `x(t)`, Fenster `w(t)` und Referenzschwingungen `cos(2pi f t)` bzw. `-sin(2pi f t)`.
2. **Produkte links unten**: die Produkte `x(t)w(t)cos(2pi f t)` und `x(t)w(t)(-sin(2pi f t))` samt Integralwert.
3. **Komplexe Ebene in der Mitte**: Zusammensetzen von Real- und Imaginärteil zu einem Zeiger.
4. **Spektrum rechts**: Betrag und Phase des Fourier-Koeffizienten als Funktion der Frequenz.

### Bedienung

- **Signalwahl**: Mischsignal, Cosinus, Sinus oder konstantes Signal
- **Fensterwahl**: Rechteck-, Hann- oder Hamming-Fenster
- **Observation duration**: Länge des betrachteten Zeitfensters
- **Signal frequency**: Frequenz des Testsignals bei Cosinus oder Sinus
- **Probe frequency**: Frequenz, für die der Fourier-Koeffizient gerade berechnet wird

### Hinweise zur Interpretation

- Wenn die Probe frequency auf einer vorhandenen Signalfrequenz liegt, wird der Zeiger typischerweise lang.
- Die Orientierung des Zeigers zeigt die Phasenlage.
- Beim Wechsel der Fensterfunktion verändern sich Peakbreite und Nebenkeulen.
- Beim Variieren von `T_obs` lässt sich die spektrale Auflösung direkt beobachten.
